```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
iris = snr.load_dataset('iris')
```

```python
iris.head()
```

```python
ax = iris.plot(figsize = (12,9),title = 'visulization of iris')
```

```python
ax.set_xlabel('axis = 1')
```

```python
ax.set_ylabel('axis = 1')
```

```python
data = iris.drop(['species'], axis =1)
```

```python
data.head()
```

```python
data[6:11].plot(kind = 'bar')
```

```python
data[10:15].plot(kind = 'barh')
```

```python
d = data.iloc[100:105]
```

```python
d.plot.pie(subplots = True,figsize = (10,10),fontsize = 10, autopct = '%.2f')
```

```python
data.diff().plot(kind = 'hist', bins = 200)
```

```python
data['sepal_width'].diff().plot(kind = 'hist', bins=200)
```

```python
data['petal_length'].diff().plot(kind = 'hist', bins = 200)
```

```python
data.plot.scatter(x = 'petal_length', y = 'petal_width', c = 'sepal_width')
```

```python
new_data = pd.read_csv('/content/tips.csv')
```

```python
new_data.head()
```

```python
snr.relplot(x = 'total_bill', y = 'tip',hue = 'sex', size = 'size',style = 'time',col = 'day' ,data = new_data)
```

```python
snr.catplot(x = 'total_bill', y = 'day', data = new_data)
```

```python
snr.catplot(x = 'tip', y = 'sex', kind = 'box', data = new_data)
```

```python
snr.pairplot(x_vars = ['total_bill', 'tip', 'size'], y_vars = 'day', data = new_data)
```

```python
plt.figure(figsize = (10,10))
```

```python
snr.heatmap(new_data.corr(), annot = True)
```
